```python
with open("notes.txt","w") as file:
    file.write("RAG = Retrieval Augmented Generation\n")
    file.write("API = Application Programming Interface\n")
    file.write("LLM = Large Language Model\n")
    file.write("Python = Programming Language for AI\n")
    file.write("Vector DB = Database used for Embeddings\n")


def load_notes(file_name):
    notes = {}

    with open(file_name, "r") as file:
        for line in file:
            if "=" in line:
                key, value = line.strip().split("=", 1)
                notes[key.strip().lower()] = value.strip()

    return notes


def search_notes(notes, topic):
    return notes.get(topic.lower(), "Topic not found")


def add_note(filename):
    key = input("Enter topic: ")
    value = input("Enter meaning: ")

    with open(filename, "a") as file:
        file.write(f"{key} = {value}\n")

    print("Note added successfully!")


notes_file = "notes.txt"
notes = load_notes(notes_file)

while True:
    print("\n1. Search note")
    print("2. Add note")
    print("3. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        topic = input("Enter topic: ")
        print(search_notes(notes, topic))

    elif choice == "2":
        add_note(notes_file)
        notes = load_notes(notes_file)   # refresh dictionary

    elif choice == "3":
        print("Goodbye!")
        break

    else:
        print("Invalid input")
```
