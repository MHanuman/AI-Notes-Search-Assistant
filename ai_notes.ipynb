```python
with open("notes.txt","w") as file:
    file.write("RAG = Retrieval Augmented Generation\n")
    file.write("API = Application Programming Interface\n")
    file.write("LLM = Large Language Model\n")
    file.write("Python = Programming Language for AI\n")
    file.write("Vector DB = Database used for Embeddings")

#to convert the data into key value pairs
def load_notes(file_name):
    notes={}
    
    with open(file_name,"r") as file:
        for line in file:
            if "=" in line:
                key,value=line.strip().split("=",1)
                notes[key.strip().lower()]=value.strip()
    return notes

#To search value for given key 
def search_notes(notes,topic):
    topic=topic.lower()
    return notes.get(topic,"Topic not found")

notes=load_notes("notes.txt")

while True:
    topic=input("Enter Topic(or exit):")
    
    if topic.lower() == "exit":
        break
    print(search_notes(notes,topic))
```
